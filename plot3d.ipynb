### 3D Plot

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
%matplotlib notebook
fig = plt.figure()
ax = Axes3D(fig)

x = np.arange(0,10)
y = np.arange(0,10)
z1 = 100 - np.arange(-10,10)**2
y1 = np.arange(-10,10)

ax.plot(x,y,0)
for i,yp in enumerate(y1):
    ax.plot([3,3],[yp,yp],[0,z1[i]],color='red')
```
